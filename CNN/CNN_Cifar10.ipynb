```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
num_classes = 10
(x_train,y_train), (x_test, y_test)=tf.keras.datasets.cifar10.load_data()
x_train, x_test = x_train/255.0, x_test/255.0
print(x_train.shape)
print(x_train[0].shape)
print(y_train.shape)
```

```python
y_train = y_train.reshape(-1)
y_test = y_test.reshape(-1)
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
plt.imshow(x_train[4])
plt.xlabel(class_names[y_train[4]])
plt.show
```

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(32, (3,3), input_shape=(32,32,3),padding='Same',activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(tf.keras.layers.Conv2D(64,(3,3), padding='Same', activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(tf.keras.layers.Conv2D(64,(3,3), padding='Same',activation='relu'))

model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64,activation='relu'))
model.add(tf.keras.layers.Dense(num_classes, activation='softmax'))

model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
def generator(x_train, y_train, batch_size):
  while True:
    x_train, y_train = shuffle(x_train, y_train)

    for start in range(0, x_train.shape[0], batch_size):
      x_batch = x_train[start: start + batch_size]
      y_batch = y_train[start: start + batch_size]
      yield x_batch, y_batch
```

```python
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath='weights-best-file.hdf5', 
                                                monitor = 'accuracy', 
                                                verbose = 1, 
                                                save_best_only=True,)

callback = [checkpoint]
```

```python
history = model.fit(generator(x_train, y_train,1000) ,epochs=40, batch_size=1000, steps_per_epoch=len(x_train)//1000, callbacks=callback)
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['accuracy'], label='accuracy')
plt.legend()
plt.show()
```

```python
model.evaluate(x_test, y_test)
y_predicted = model.predict(x_test)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
for i in range(10):
    plt.imshow(x_test[i], 'gray')
    plt.xlabel(class_names[y_predicted_labels[i]])
    print("y_label",class_names[y_test[i]])
    plt.show()
```
