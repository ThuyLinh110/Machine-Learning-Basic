```python
import matplotlib.pyplot as plt
import cv2
import numpy as np 
import scipy
```

```python
image = plt.imread('/content/drive/MyDrive/AI_SGROUP/Machine_Learning/Segmentation/Valve_original_(1).png')
plt.imshow(image)
```

```python
# gray = rgb2gray(image)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# defining the sobel filters
sobel_horizontal = np.array([np.array([1, 2, 1]), np.array([0, 0, 0]), np.array([-1, -2, -1])])
print(sobel_horizontal, 'is a kernel for detecting horizontal edges')
 
sobel_vertical = np.array([np.array([-1, 0, 1]), np.array([-2, 0, 2]), np.array([-1, 0, 1])])
print(sobel_vertical, 'is a kernel for detecting vertical edges')
```

```python
out_h = scipy.ndimage.convolve(gray, sobel_horizontal, mode='reflect')
out_v = scipy.ndimage.convolve(gray, sobel_vertical, mode='reflect')
```

```python
plt.imshow(out_h, cmap='gray')
```
