```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt

dataset=pd.read_csv("/content/drive/MyDrive/AI_SGROUP/Machine_Learning/Linear Regression/data.csv")
x=dataset.iloc[:, [2,3,4,5,6,7,8,9]].values
x=x.T
y=dataset.iloc[:, [1]].values

x_train =x[:,:3000]
y_train =y[:3000]
x_test = x[:,3000:]
y_test = y[3000:]


m=len(x_train)  # so chieu cua input 8
n=len(x_train[0]) # so sample 3000
w = np.ones((m, 1))
b = 1

def loss_derivative(w, b, x, y):
    n = y.shape[0]
    delta = (np.dot(w.T, x) + b).T - y     # y^ -y
    loss = np.sum(0.5 * ((delta)**2)) / n  # loss = 1/2 (y^ - y )^2
    dw = np.dot(x, delta) / n
    db = np.sum(delta) / n
    return loss, dw, db


epoch = 3000 # so lan lap
lr = 0.000000001
loss, dw, db=loss_derivative(w,b,x_train, y_train)
l=loss
for i in range(epoch):
    loss, dw, db=loss_derivative(w,b,x_train, y_train)
    # if loss<l:
    #   break
    w -= lr * dw
    b -= lr*db
    l=loss
for i in range(30):
    print("y_linear_regression: ",np.dot( x_test[:,i].T, w) +b  )
    print("y_test             : ",y_test[i])
    print("_________________________________________________")

```
